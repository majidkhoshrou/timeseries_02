```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
from sklearn import mixture
# https://machinelearningmastery.com/time-series-data-visualization-with-python/
```

```python
df = pd.read_csv('data/philips.csv', skiprows=1, parse_dates=['Week'], index_col='Week')
assert all(df.sort_index().index == df.index)
df.columns = ['philips_Australia']
df.head(3)
df = df.loc['2015':'2019']
```

```python
df.plot();
plt.xlabel("year")
```

```python
df2 = df.loc['2016':'2019']
df2 = df2.resample('W').mean()
df2.index
groups = df2.groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = pd.Series(group.philips_Australia.values)
#     print((pd.Series(group.philips_Australia)))
#     print((pd.Series(group.philips_Australia)))
years = years.T
years
plt.matshow(years, interpolation=None, aspect='auto')
plt.colorbar()
plt.show()
```

```python
X = df2[["philips_Australia"]]
```

```python
clf = mixture.GaussianMixture(n_components=5, covariance_type='full')
clf.fit(X)
```

```python
y = clf.score_samples(X)
y
```

```python
idx = X.index[abs(y)>5.9]
idx
```

```python
plt.plot(df2)
plt.plot(idx, X.loc[idx,:], marker='.', linestyle = 'None')
plt.savefig('anomaly01.pdf')
```
